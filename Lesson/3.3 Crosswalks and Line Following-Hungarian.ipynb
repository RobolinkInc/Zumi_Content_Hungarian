## Vonalkövetés
<font size=3> Most ahelyett, hogy Zümi megállna, amikor meglátja a vonalat, követni fogja. Útközben Zümi a kamerával érzékeli a sávokat, és rajta marad a vonalon. Ez arra készteti Zümit, hogy a vonalon maradjon egy függvény hívásával. Először ismerkedjünk meg a működésével.
    
<img src="../Data/images/line_follower.png" width=400>

Mivel ez a lecke az alsó IR érzékelőkről szól, ez a kód mindkettőt használni fogja a vonal ellenőrzésére. Tegyél úgy, mintha egy darab szalagon járnál a földön. Ha a jobb lábad lelép a vonalról, melyik irányba fordulsz? Mi a helyzet a bal lábaddal? Az infravörös érzékelők ugyanezt a logikát használják. <br> <br>
    
<img src="../Data/images/black_line.png" width=400></font>

### Line_follower()
<font size=3> Teszteljük a függvényt! Használj szigetelőszalagot vagy az autósiskola szőnyeged (ha van), és állítsd Zümit a kezdő fekete vonalra. A <font face="Courier">zumi.line_follower()</font> függvénynek legalább egy paraméterre van szüksége, ami nem más, mint az időtartam. Például futtassuk a programot 3 másodpercig.

```python
zumi.line_follower(3)
```

<font size=3> Ha úgy gondolod, hogy a Zümi nem nagyon érzékeli a vonalakat, vagy hogy a sötét és a világos közötti különbség nem elég nagy, akkor lehet, hogy módosítanod kell azokat a **határokat** vagy határértékeket, amelyeket az IR érzékelők észlelni fognak. Alapértelmezés szerint 100-ra van állítva. Az IR-érzékelők teszteléséhez és a küszöbértékek meghatározásához lásd a 3.1. leckét.</font><br>

<font size=3><font face="Courier">zumi.line_follower(3, left_thresh=100,right_thresh=100)</font>

<font size=3>**Megjegyzés**: Ez a függvény egy while ismétlődő parancssort hív. Ha mindkét infravörös érzékelő fehéret lát (a sor vége), akkor a kód automatikusan leáll, még akkor is, ha az idő nem telt le.</font>
