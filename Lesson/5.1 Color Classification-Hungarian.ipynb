# Hogyan látják és értelmezik a számítógépek a színeket?
<font size =3> Mielőtt elkezdenéd a Zümi tanítását a színek felismerésére, meg kell tanulnod, hogyan látja Zümi a színeket. Ez nagyon különbözik attól, ahogy te látod a színeket!

## Mi az a kép?
<font size =3> A kép **pixelekből** (képelem rövidítése), kis pontok tömbjéből áll. A pixel lehet egyetlen szín vagy színkombináció, és ezeket a színeket három szám sorozata képviseli, amelyek pontosan megmondják, hogy mennyi benne a piros, zöld és kék szín. Ezt hívják **RGB** értéknek, amely piros, zöld és kék színből áll. Például egy gyönyörű türkiz árnyalat száma (27, 209, 197), mivel nincs benne sok piros, de sok benne a zöld és a kék. Mivel az RGB minden értéke 0 és 255 között lehet, az egyes színekhez 256 érték közül lehet választani. Tehát 256^3, vagyis 16 777 216 különböző színkombináció jöhet létre! </font>


## Mi az a mátrix?
<font size =3> Mivel minden egyes képpont számokkal ábrázolható, a kép számrácsból áll. Itt kezdik az emberek és a számítógépek kicsit másképp látni a képeket. Az emberek olyan színeket és alakzatokat látnak, amelyeket különböző tárgyakként ismerünk fel, de a számítógépek csak ezeket a rácsokat, más néven **mátrixokat** látják. Minden szám az egyes pixelek RGB-értékét jelenti. Valahogy így néznek ki: </font>

<img src="../Data/images/matriceszumi.png" width=500> 

## HSV használata RGB helyett

<font size =3> 

A program minden RGB képet **HSV** színtérré alakít. A HSV jelentése színárnyalat, telítettség és érték.


* **Hue:** általában 0-360 közt van és a színt jelöli (ebben az alkalmazásban azonban 0-180)
* **Saturation:** a szín intenzitása
* **Value:** azt mutatja, hogy mennyire világos vagy sötét a szín

A számítógépes látást biztosító alkalmazásokban jobb a HSV színtér használata, mivel ez elválasztja a színek és az intenzitás értékeit. Ez azért fontos, mert az árnyékok, a tükröződések és más tényezők hatására egyes színek nagyon eltérőek lehetnek. A HSV színtér ezt figyelembe veszi a pontosabb eredmények érdekében. </font>

## Modell betöltése
<font size=3> A gépi tanulás nagyszerű jellemzője, hogy amikor elmentesz egy modellt, újra felhasználhatod azt. Az előző tevékenység során a három kedvenc színedet választottad ki, hogy tanítsd Zümit. Most Python kóddal fogod betölteni a modellt, és Zümi minden egyes színre másképp fog reagálni.</font>
    
### Könyvtárak importálása
<font size=3> Importáld a következő könyvtárakat a vezetéshez, a fényképezéshez és a színosztályozáshoz szükséges függvények eléréséhez.

```python
from zumi.zumi import Zumi
from zumi.util.camera import Camera
from zumi.util.screen import Screen
from zumi.util.color_classifier import ColorClassifier
import time

camera = Camera()
screen = Screen()
zumi = Zumi()
```

### Modell betöltése
<font size=3> A modell betöltéséhez hívj meg egy függvényt a ColorClassifier könyvtárból: <font face="Courier">load_model()</font>. Ehhez a függvényhez szükséged van egy paraméterre, ami a projekt neve. Emlékszel hogy nevezted el a projekted? Írd be a felhasználóneved és a projekt nevét az alábbi idézőjelek közé. Ellenőrizd, hogy a helyesírás és az írásjelek helyesek. Ha nem létezik, akkor hibaüzenetet kapsz.

```python
user_name = ''
demo_name = ''
```

```python
knn = ColorClassifier(user_name=user_name)
train = knn.load_model(demo_name)
knn.fit("hsv")
```

<font size=3> Miután sikeresen betöltötted a modelled, tesztelheted. Futtasd az alábbi kódot, és válassz egy színkártyát, amelyet megmutatsz Züminek. Ha készen állsz, nyomd meg az Enter billentyűt az eredmények megjelenítéséhez a képernyőn. Ha megnyomod a "q" gombot, a program kitör az ismétlődő parancssorból, és kikapcsolja a kamerát.

```python
camera.start_camera()

while True:
        user_input = input("Press 'enter' to predict or 'q to quit: ")

        if user_input == "q":
            break
            
        image = camera.capture()
        predict = knn.predict(image)
        screen.draw_text_center(predict)

camera.close()
```
