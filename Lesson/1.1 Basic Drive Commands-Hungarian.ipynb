## Meghajt√≥parancsok

<font size =3> Z√ºmi k√ºl√∂nb√∂z≈ë f√ºggv√©nyeket haszn√°l a vezet√©shez. Ebben a leck√©ben r√©szletesebben elmagyar√°zzuk, milyen f√ºggv√©nyek vannak, mit csin√°lnak, √©s hogyan haszn√°lhatod ≈ëket a t√°vir√°ny√≠t√≥d l√©trehoz√°s√°hoz. </font>


### Mik a f√ºggv√©nyek?
<font size =3> Ahhoz, hogy a Z√ºmi elinduljon, n√©h√°ny **f√ºggv√©nyt** kell haszn√°lnunk. Gondolj √∫gy a f√ºggv√©nyekre, mint egy k√≥dcsomagokra, amelyek seg√≠ts√©g√©vel hat√©konyabb√° teheted a programod. Lehetnek **bemenetei** √©s **kimenetei**.</font>

### Z√ºmi f√ºggv√©nyei
<font size=3>Az al√°bbiakban felsoroljuk az alapvet≈ë hajt√°sfunkci√≥kat:

* <font face="Courier">forward()</font>: Elindul el≈ëre 40-es sebess√©ggel 1 m√°sodpercig abba az ir√°nyba, amerre Z√ºmi n√©z 
* <font face="Courier">reverse()</font>: Elindul h√°tra 40-es sebess√©ggel 1 m√°sodpercig abb√≥l az ir√°nyb√≥l, amerr≈ël Z√ºmi n√©z
* <font face="Courier">turn_left()</font>: 90 fokkal balra fordul
* <font face="Courier">turn_right()</font>: 90 fokkal jobbra fordul

</font>

### Hogyan kezdj√ºk a f√ºggv√©nyeket?
<font size =3> A sz√°m√≠t√°stechnik√°ban valaki felsz√≥l√≠t√°sa, alapvet≈ëen a futtat√°s√°t jelenti. A f√ºggv√©nyeket a t√°rgy nev√©vel kell elnevezni, amely jelen esetben a <font face="Courier">Z√ºmi</font>. 
    
<img src="../Data/images/zumi_forward.jpg" width=700>   
    
Az al√°bbi cell√°ban van egy p√©lda a <font face="Courier">forward()</font> f√ºggv√©nyre. A Z√ºmi egy m√°sodpercig halad majd el≈ëre, ez√©rt √ºgyelj arra, hogy elegend≈ë hely √°lljon rendelkez√©s√©re a ter√ºleten! </font>


```python
zumi.forward()
```

<font size =3>  Most pr√≥b√°ljuk meg h√°tra... </font>

```python
# ITT √çrj egy k√≥dot, hogy Z√ºmi 1 m√°sodpercig h√°trafel√© haladjon
```

<font size=3>N√©zz√ºk √°t a k√∂vetkez≈ë k√©t f√ºggv√©nyt. A <font face="Courier">turn_left()</font> √©s <font face="Courier">turn_right()</font> felsz√≥l√≠t√°s Z√ºmi balra vagy jobbra kanyarod√°s√°hoz vezet.

Teszteld ezt a k√≥dot lent, majd adj hozz√° tov√°bbi parancsokat b√°rmilyen sorrendben, hogy l√°sd, mikor mi t√∂rt√©nik. Ha szeretn√©l egy kis id≈ët az egyes parancsok k√∂z√∂tt, adj meg egy <font face="Courier">time.sleep(seconds)</font>  parancsot, hogy a programot a megadott m√°sodpercekig k√©sleltesse.
Futtasd az al√°bbi k√≥dot, hogy l√°sd, hogyan m≈±k√∂dik, majd pr√≥b√°lj meg hozz√°adni m√©g n√©h√°ny parancsot a k√≥dhoz. </font>

```python
zumi.forward() # Menjen el≈ëre 1 m√°sodpercig, majd √°lljon meg
time.sleep(2) # V√°rjon 2 m√°sodpercet
zumi.turn_right() # Forduljon jobbra 90 fokot
# Adj hozz√° m√©g n√©h√°ny k√≥dot!
```
