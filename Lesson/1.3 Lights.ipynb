## Mi az a LED?

<font size =3> A LED a "light-emitting diode" rövidítése. A dióda olyan elektromos eszköz, amely lehetővé teszi az elektromos áram egyetlen irányban történő áramlását. A LED-ek már lecserélik a hagyományos izzókat, mert hatékonyabbak. </font>

### Zümi LED-jei
<font size =3>A Zümi tábláján 8 LED található, de nem lehet mindet beprogramozni. Három LED van a Zümi bal oldalán, amelyek jelzik a töltési állapotot, és egy fehér LED a Zümi jobb oldalán, amely azt jelzi, hogy a Pi sikeresen elindult. A tábla elején két LED található a fényszórók számára. Ne keverd össze őket az infravörös (IR) érzékelőkkel! Később megismerheted ezeket is. A fényszórókat a Zümi külseje rejti. </font>

## Zümi LED függvényei

<font size =3> A fényszórók és a féklámpák be- és kikapcsolása mellett irányjelzőket is használhatsz.
Itt található az összes használható függvény:

```
* all_lights_on()

* all_lights_off()

* headlights_on()

* headlights_off()
       
* brake_lights_on()
       
* brake_lights_off()
        
* hazard_lights_on()
        
* hazard_lights_off()

* signal_left_on()
       
* signal_left_off()
       
* signal_right_on()
       
* signal_right_off()
```     

Most próbáld meg beírni az irányjelző függvényeket a távirányítód kódjába, vagy írj néhány új kódot.
Itt van egy példa, amelyet kipróbálhatsz. Győződj meg róla, hogy sok helyed van ehhez.
</font>

````python
zumi.signal_right_on()
zumi.forward(duration=3)
zumi.turn_right()
zumi.signal_right_off()
zumi.forward()
````
