### Függvények meghatározása
<font size=3> A saját függvény megírásának első lépése a <font face="Courier">def</font> kulcsszó. A Python tudomására hozod, hogy egy függvényt fog **definiálni**. Ezután add meg a függvény nevét, amelyet zárójel és kettőspont követ. A kettőspont után behúzás következik, ami tartalmazza a tartalmazza a függvény törzsét, ami magában foglalja azt, amit a függvényében meg szeretnél tenni. Hozzunk létre például egy olyan függvényt, amely üdvözöl!</font>

```python
def greet_user():
    print("Hello")
```

<font size=3>Ha a fenti kódot futtatod, semmi sem fog történni. Ez a kód csak meghatározza a függvényt, de addig nem fut, amíg nem szólítod fel. </font>

```python
greet_user() # Szólítsd fel a függvényt!
```

### Függvények meghatározása paraméterekkel
<font size=3> Mi van akkor, ha azt akarjuk, hogy a függvény a megadott paraméterektől függően mást csináljon? Ne feledd, hogy megváltoztathatod a Zümi előre haladásának másodpercét a <font face="Courier">zumi.forward()</font> fügvénnyel. Paramétereket is hozzáadhatsz a függvényeidhez! </font>

```python
def greet_user(name):
    print("Hello", name)

greet_user("Zumi")
```

<font size=3> Módosítottuk a függvénydefiníciót úgy, hogy "helló" legyen a paraméterként megadott név. A példában "Zümi"-t **adunk meg** paraméterként a függvénynek. Próbáld újra felszólítani a függvényt, de más néven. Ügyelj arra, hogy idézőjelben szerepeljen! </font>

```python
greet_user("VÁLTOZTASS MEG!")
```

<font size=3> Szerinted mi történne, ha a függvényhívásban nem lennének idézőjelek? Vagy, ha számokat teszünk oda?  </font>

```python
greet_user(12345)
```

<font size=3> Mivel már meghatároztuk a képernyőre írás függvényét, minden bevitelt idézőjelben kell megadni. Ez nem azt jelenti, hogy nem írhatsz olyan függvényeket, amelyek számokat használnak. Zümi paramétereinek többsége, például a sebesség és az időtartam, számokból állnak. </font>
