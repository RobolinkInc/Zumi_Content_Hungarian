## Ha állítások
<font size =3>A Pseudocode-ban a programodnak utasítania kell Zümit, hogy haladjon egy adott irányba a bemenet alapján. "Ha" utasításokat vagy **feltételeket** használ, hogy segítse a programot a döntések meghozatalában. Valójában minden nap használsz feltételes mondatokat. Néhány példa erre:


* Ha az ébresztőm megszólal, akkor felébredek. Egyébként elalszom.
* Ha éhes vagyok, eszem valamit.
* Ha kint süt a nap, fényvédőt fogok viselni.
* Ha tiszta a szobám, megkapom a zsebpénzemet. Egyébként nem kapok pénzt.


A Phyton-ban így írunk "ha" utasításokat: <br> <br>

<font face="Courier">
    
if insert condition here: <br>
<span style="margin-left: 40px;">Insert statements here</span> <br>
else: <br>
<span style="margin-left: 40px;">Insert statements here</span> <br>

if alarm goes off:<br>
<span style="margin-left: 40px;">I will wake up</span><br>
else:<br>
<span style="margin-left: 40px;">Stay asleep</span><br>
</font>

<img src="../Data/images/conditional_flowchart.jpg" width=700>



A számítógép a feltételt ellenőrzi, hogy igaz-e vagy hamis.
* Ha a feltétel igaz, a számítógép végrehajtja vagy kivitelezi az alatta lévő utasításokat.
* Ha a feltétel hamis, a számítógép NEM hajtja végre az alatta lévő utasításokat.
* Az "Else" parancsnak nincs feltétele! Ez azért van, mert az "else" az utolsó lehetőség az if-else utasítások használatakor. Amikor a számítógép eléri az "else" opciót az összes többi feltétel ellenőrzése után, akkor az else feltétel automatikusan igaz és végrehajtja az utasításokat.
* Az "else" használatának egyetlen módja az, ha valaminek történnie kell, és a feltétel hamis. Ha a "ha" feltétel hamis, és semminek sem kell történnie, ne használj "else" parancsot.

## Összehasonlítók
<font size =3> Ezután fordítsa le a Pseudocode-od Python-kódra. Mielőtt meghajtóparancsokat adnál meg, használj írási utasításokat a teszteléshez, mint például ez: <br> <br>
    
<font face="Courier"> 
direction = input("Please enter a command: ")

if direction == "w": <br>
<span style="margin-left: 40px;">print("Go forward")</span> <br>    
</font>
    
Az egyenlőségjel ismerősnek tűnhet, de a kódban kissé más az alkalmazása. Az <font face="Courier">==</font> jel egy **összehasonlító** példa. Az összehasonlítók belső körülmények között használhatók az értékek összehasonlítására. Ha az állítás igaz, akkor az az alatti állítás végrehajtásra kerül. Ezek összehasonlító jelek:

* \> Nagyobb, mint
* \< Kisebb, mint
* \>= Nagyobb vagy egyenlő
* <= Kisebb vagy egyenlő
* == Egyenlő
* != Nem egyenlő

Futtasd az alábbi cellát, és végezz néhány módosítást a feltételekben, hogy lásd, hogyan működnek ezek. Ha akarod, megváltoztathatod a változó nevét vagy a bemeneti utasításokat. Ellenőrizd a kódot.
</font>

```python
X= 10
Y = 100
Z = 200

if X != Y:
    print("X is not equal to Y")

if Z > Y:
    print("Z is greater than Y")
```

<font size =3> Most már megvan az összes eszköz, amellyel kiírhatod a hiányzó feltételeket. Használd a "ha" utasításokat, az összehasonlítókat és írd ki az utasításokat, majd írj kódot az "s", "a" és "d" billentyűkhöz.

```python
#ITT Fejezd be a hiányzó if utasítások többi részét

while True:
    direction = input("Please enter a command: ")

    if direction == "w":
        print("Go forward")
```

<font size =3> Észrevetted, hogy a kód örökké fut? Mivel ez egy távirányító kód, azt szeretnéd, ha folyamatosan megmondhatnád Züminek, hogy mit kell tennie. Ehhez egy örökké futó "while" ciklust kell használnod. Jelenleg a kód csak akkor állítható le, ha az eszköztáron a "leállítás" ikonra kattint. Ahhoz, hogy biztonságosan kiléphess ebből a ciklusról, jobb, ha a kód tartalmaz egy if utasítást, amely észlelni képes egy "q" bemenetet, hogy ki tudj szállni a ciklusból. <br> <br>
    

<font face="Courier">
if direction =="q": <br>
<span style="margin-left: 40px;">break</span>
</font>

Add hozzá ezt a kódot a fenti kódodhoz, és teszteld, hogy hiba nélkül kiléphetsz-e a kódból. </font>
