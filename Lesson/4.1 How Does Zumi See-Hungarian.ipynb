### Sz√ºrke√°rnyalat
<font size=3> A sz√ºrke√°rnyalatot szok√°s "fekete-feh√©rnek" nevezni. Ez azonban nem igaz√°n pontos, mert a k√©p sz√ºrke k√©ppontokb√≥l √°ll. A sz√ºrke√°rnyalatos k√©peket gyorsabban lehet feldolgozni, mert nincsenek m√°s sz√≠nek benne. K√©s≈ëbb sz√ºrke√°rnyalatos k√©peket fogsz haszn√°lni a QR-k√≥dok beolvas√°s√°hoz!

```python
camera.start_camera()
img = camera.capture()
camera.close()
gray = vision.convert_to_gray(img) # Sz√ºrk√©v√© konvert√°l√°s
camera.show_image(gray)
```

### HSV 
<font size=3> A HSV a **sz√≠n√°rnyalat**, a **tel√≠tetts√©g** √©s az **√©rt√©k**. Annak ellen√©re, hogy a k√©p furcs√°nak t≈±nik sz√°modra, ez a sz√≠nt√©r akkor hasznos, amikor Z√ºminek fel kell fedeznie vagy nyomon kell k√∂vetnie bizonyos sz√≠neket. Hasznosabb, mint a szok√°sos sz√≠nes k√©p, amelyet l√°tni szokt√°l, mert minden inform√°ci√≥ pixel meg tudja mondani a sz√°m√≠t√≥g√©pnek a sz√≠n intenzit√°s√°t √©s azt, hogy vannak-e benne √°rny√©kok.

```python
camera.start_camera()
img = camera.capture()
camera.close()
hsv = vision.convert_to_hsv(img) # Konvert√°ld HSV-re, sz√≠n√°rnyalatra, tel√≠tetts√©gre √©s √©rt√©kre
camera.show_image(hsv)
```

### Ford√≠tva
<font size=3> Ezt csak sz√≥rakoz√°sb√≥l! Ez a sz≈±r≈ë megford√≠tja a sz√≠n √°rnyalatait. P√©ld√°ul a vil√°gosabb ter√ºletek s√∂t√©tebbek, a s√∂t√©tebb ter√ºletek pedig vil√°gosabbak lesznek.</font>

```python
camera.start_camera()
frame = camera.capture()
camera.close()
invert = cv2.bitwise_not(frame) # sz√≠nek megford√≠t√°sa
camera.show_image(invert)
```

## Vide√≥

<font size =3> B√°r egy vide√≥ sim√°nak n√©z ki, val√≥j√°ban egym√°s ut√°ni k√©psor. A k√©pek olyan gyorsan jelennek meg, hogy √°ltal√°ban nem √©szlelsz k√ºl√∂nbs√©get. Itt azonban √©szreveheted a k√ºl√∂nbs√©get, k√ºl√∂n√∂sen, ha a k√©peid nagyon nagyok. A vide√≥ megjelen√≠t√©s√©hez k√©sz√≠ts √©s jelen√≠ts meg k√©peket egy for ism√©tl≈ëd≈ë parancssoron bel√ºl.
<font size =3>A vide√≥ megjelen√≠t√©s√©hez t√∂ltsd ki a lenti k√≥dot. Mivel ism√©tl≈ëd≈ë parancssort fogsz haszn√°lni, bevezet√ºnk k√©t √∫j k√≥dr√©szt, amelyek seg√≠tenek megakad√°lyozni a program √∂sszeoml√°s√°t: <font face="Courier">try</font> √©s <font face="Courier">finally</font>.

```python
camera = Camera()
camera.start_camera()

try:
    for x in range(30):
        # K√©sz√≠ts egy k√©pet
        # Jelen√≠tsd meg a k√©pet
        camera.clear_output() # T√∂r√∂ld a k√∂vetkez≈ë megjelen√≠tend≈ë k√©p kimenet√©t
finally:
    camera.close()
```

<font size=3> Ha b√°rmi baj t√∂rt√©nik, vagy le√°ll√≠tod a k√≥dot a <font face="Courier">try</font> szakaszban a program automatikusan a <font face="Courier">finally</font> √°ll√≠t√°sra l√©p. Ebben az esetben egy <font face="Courier">close()</font> f√ºggv√©nyt adunk hozz√° a f√©nyk√©pez≈ëg√©phez, hogy soha ne kelljen att√≥l tartanod, hogy bekapcsolva marad. </font>
