## Az IR adatok megjelenítése a képernyőn
<font size =3>Most meg fogod jeleníteni az IR-adatokat a Zümi képernyőn. Az IR-adatok megszerzéséhez szükséged lesz egy Zümi objektumra, és a képernyőn való megjelenítéshez létre kell hoznod egy képernyő objektumot. </font>

### Könyvtárak importálása

```python
from zumi.zumi import Zumi
from zumi.util.screen import Screen
import time

zumi = Zumi()
screen = Screen()
```

### Mit jelentenek az adatok?

<font size =3> Az IR szenzorok úgy működnek, hogy infravörös impulzust bocsátanak ki, és mérik az infravörös fényt, amely visszatér, miután lepattan egy tárgyról. Ez a szám 0 és 255 között lesz.
Az alacsonyabb szám azt jelzi, hogy a közelben lévő tárgy visszatükrözi az infravörös fényt az érzékelő felé.

Futtasd az egyes cellákat, és figyeld a képernyőn megjelenő számok változását, miközben lefedi az érzékelőket. A kód a for ismétlődő parancssort használja 50 leolvasáshoz. </font>

#### Első IR szenzoros

```python
for i in range(0,50):
    ir_readings = zumi.get_all_IR_data()
    front_right_ir = ir_readings[0]
    front_left_ir = ir_readings[5]
    
    message = "    IR readings        "
    message = message + str(front_right_ir) + ", " + str(front_left_ir)
    screen.draw_text(message)
    time.sleep(0.1)
screen.draw_text_center("Done!")
```

#### Hátsó IR szenzorok


```python
for i in range(0,50):
    ir_readings = zumi.get_all_IR_data()
    back_right_ir = ir_readings[2]
    back_left_ir = ir_readings[4]
    
    message = "    IR readings        "
    message = message + str(back_right_ir) + ", " + str(back_left_ir)
    screen.draw_text(message)
    time.sleep(0.1)
screen.draw_text_center("Done!")
```

#### Alsó IR szenzorok

<font size =3> Az alsó infravörös érzékelők kiválóan alkalmasak vonalkövetésre vagy észlelésre. Használd az alsó IR munkalapot a világos és a sötét értékek teszteléséhez.</font>

```python
for i in range(0,50):
    ir_readings = zumi.get_all_IR_data()
    bottom_right_ir = ir_readings[1]
    bottom_left_ir = ir_readings[3]
    
    message = "    IR readings        "
    message = message + str(bottom_right_ir) + ", " + str(bottom_left_ir)
    screen.draw_text(message)
    time.sleep(0.1)
screen.draw_text_center("Done!")
```

## Listák

<font size=3> Talán észrevehettél valamit az IR-adatokról, amit még eddig nem tanultál. Az összes példában a <font face="Courier">get_all_IR_data()</font> függvényt használjuk. Van azonban hat IR-érzékelő, amelyek mindegyikének van értéke. Hogyan lehetnek mind ugyanabban a változóban? Ezeket az adatokat egy **lista** tárolja. A Pythonban a listák felhasználhatók azonos típusú adatok tárolására és rendszerezésére több változó létrehozása nélkül. </font> 

### Hozzáférés az értékekhez

<font size=3> Amikor a <font face="Courier">zumi.get_all_IR_data()</font> függvényt használod, a program kéri az összes szenzor adatot, és elmenti őket egy listába. Ebben a példában a lista neve <font face="Courier">ir_readings</font>. Ellentétben az általában függőleges bevásárló listákkal, a Pythonban található listát vízszintesen írják és [zárójelbe] zárják. A következő cella leolvassa és kiírja az értékeket.

```python
ir_readings = zumi.get_all_IR_data()
print(ir_readings)
    
```

<font size=3> A lista nevének kiírásával az összes érték ki lesz írva, de mi van, ha csak egy vagy két értéket szeretnél ellenőrizni? Az adatokhoz vagy az egyes **elemekhez** azok indexe alapján férhetsz hozzá. Az index olyan, mint a lakcím, ahol az elem lakik a listában. Az ismétlődő parancssorhoz hasonlóan a Python-listák is 0-tól kezdődnek. Vegyük például ezt a listát. Mi az első elem értéke? Mi a helyzet az utolsóval?
