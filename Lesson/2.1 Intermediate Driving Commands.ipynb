### Vezetési függvények

<font size =3> Emlékszel, amikor a <font face="Courier">forward()</font> függvényről tanultál? Annak ellenére, hogy a Zümi beállította, hogy az alapértelmezett 40-es sebességgel egy másodpercig haladjon az aktuális irányban, ezeket az értékeket is megváltoztathattad. Az alább felsorolt ​​összes függvény az alapértelmezett paraméterekkel van felsorolva . Ha egy függvényt névvel szólítunk fel, például <font face="Courier">right_u_turn()</font>, Zümi a megadott paraméterekkel u-fordulatot tesz. Ugyanakkor, csakúgy, mint a <font face="Courier">forward()</font> esetében, ezeket is megváltoztathatod a kívánt értékekre. </font>
