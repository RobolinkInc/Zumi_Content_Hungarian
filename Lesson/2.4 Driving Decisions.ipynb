# Vezetési döntések

<font size =3> Ha körbenézünk észrevehetjük, hogy mindig érzékeljük a környezetünket. Képesnek kell lennünk érzékelni az út közben esetlegesen felmerülő akadályokat is. A <font face="Courier">forward()</font> fügvénnyel Zümi vezetés közben nem tudja ellenőrizni érzékelőit vagy környezetét. Képzeld el, hogy csukott szemmel járkálsz, és csak akkor nyithatod ki a szemed, ha abbahagytad a mozgást, hogy eldöntsd a következő lépésed. Ebben a leckében megtanulhatod, hogyan kell használni egy új függvényt, amely lehetővé teszi a Zümi számára a többfeladatos munkát. Ezzel képes leszel vezetni és ellenőrizni az érzékelőket a döntések meghozatalához.</font>

## go_straight()
<font size=3> Amikor sétálsz, milyen cselekedetek ismétlődnek meg? Jobb láb, bal láb, jobb láb, bal láb... Zümi is tehet "lépéseket" előre. Amikor ezeket a kis lépéseket egy ismétlődő parancssorba tesszük, Zümi előre fog mozogni. A mozgási függvény egy kis lépésnél a <font face="Courier"> zumi.go_straight()</font>. Az <font face="Courier">forward()</font> függvénytől eltérően a paraméterek nem választhatók. Züminek tudnia kell a sebességet és a **menetirányt**, vagy az irányt. Az alábbi képen mi Zümi menetiránya?
    
<img src="../Data/images/heading_straight.png" width="200">
    
Futtasd az alábbi kódot, hogy megtekinthesd ennek a vezetési mintának a kódját a <font face="Courier">go_straight()</font> függvényt használva.

```python
zumi.reset_gyro() # Minden szög nullázása a kezdéskor

for x in range(40): # Tegyél 40 lépést
    zumi.go_straight(40, -30) # Menj 40-es sebességgel a -30 menetirányból
    
zumi.stop() # Ne felejtsd el leállítani!
```

<font size=3> A <font face="Courier">stop()</font> a program végén nagyon fontos! Mivel a <font face="Courier">go_straight()</font> lehetővé teszi a gördülékeny vezetést anélkül, hogy szünetek lennének közöttük, semmi sem akadályozza Zümit abban, hogy örökké haladjon (vagy amíg az akkumulátor lemerül). A for ismétlődő parancssor befejezése után a kód folytatódik a következő sornál. Bármely kódot, amely **nem** a for ismétlődő parancssor része, nem szabad behúzni.<br><br>
A lenti cellában gyakorolhatod az ismétlődő parancssorok használatát és az új vezetési módot. Próbáld meg megváltoztatni a második fejléc-paraméterét, és becsüld meg, mennyi időbe telik egy bizonyos számú iteráció meghajtása.

### Tájékozódás
<font size=3> Ha meg szeretnéd tudni, hogy Zümi fejjel lefelé áll-e, megismerhetsz egy új függvényt: <font face="Courier">get_orientation()</font>. Hatféleképpen állhat Zümi:
    
<img src="../Data/images/orientations.png" width="750">

Az alábbi kód segítségével ellenőrizd a hat különböző állapotot. Ha 0, -1 vagy 7 értéket látsz, akkor Zümi két állapot között van, vagy az állapot ismeretlen.


```python
for i in range(80):
    zumi.update_angles()
    orientation = zumi.get_orientation() # Mentsd el ezt az értéket egy változóba, hogy később használni tudd...
    print(orientation)
    time.sleep(0.1)
    IPython.display.clear_output(wait=True) 
print("Done!")
```
